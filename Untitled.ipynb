```python
import pandas as pd
import os
import librosa as lr
import librosa.display
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
import tensorflow
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow import keras
import numpy as np
```

```python
df = pd.read_csv('./Data/features_30_sec.csv')
```

```python

res = []
for index in range(0,len(df)):     
    filename = df.iloc[index]['filename']          
    if filename == 'jazz.00054.wav':
        continue
    y, sr = librosa.load('Data/genres_original/'+filename,sr=22050)
    y_cut = y[round(0,ndigits=None)
         :round(600000, ndigits= None)]
    mfcc = librosa.feature.mfcc(y_cut,n_fft=255,hop_length=512,n_mfcc=128)
    mfccs = []
    for j in range(len(mfcc)):
        mfccs.append([np.mean(mfcc[j]),np.var(mfcc[j])])
        
    res.append(mfccs)
```

```python
res = np.array(res)
```

```python
res[:,5,0]
```

```python
df_mfccs = {}
for i in range(128):
    df_mfccs['mfcc'+str(i+1)+'_mean'] = res[:,i,0]
    df_mfccs['mfcc'+str(i+1)+'_var'] = res[:,i,1]
```

```python
len(df_mfccs['mfcc1_mean'])
```

```python
for key in df_mfccs.keys():
    df[key] = df_mfccs[key]
```

```python
cols = list(df.columns.values) #Make a list of all of the columns in the df
cols.pop(cols.index('label')) #Remove x from list
df = df[cols+['label']] #Create new dataframe with columns in the order you want
```

```python
df.to_csv('./Data/feature_extracted.csv',index=False)
```
